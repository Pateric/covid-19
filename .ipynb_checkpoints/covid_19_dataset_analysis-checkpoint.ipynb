The Severe acute respiratory syndrome coronavirus-2 (SARS-CoV-2) is responsible for the current coronavirus disease 2019 (COVID-19) pandemic. COVID-19 has now infected more than 2 million individuals and killed 120 thousand people around the world. The rapid transmission of SARS-CoV-2 continues and there is no drug or vaccine against the disease yet. While scientists are working to find the right solution, it is important to curtail the spread of this virus.

This is an Exploratory Data Analysis of the dataset which contain the case situation as at 26th April 2020 worldwide. The dataset is obtained from https://www.ecdc.europa.eu/en/publications-data/download-todays-data-geographic-distribution-covid-19-cases-worldwide. Each row/entry contains the number of new cases and deaths reported per day and per country. There are 207 countries affected by the pandemic as at 26th April from 31st Dec, 2019.

The Dataset

The dataset is a CSV file named worldwide.csv, which contains situational data of COVID-19 across
the globe from 31st Dec 2019 to 26th April, 2020. It contains the following columns:

1. dateRep: The date case (s) was recorded
2. day: Day of reporting
3. month:Month of reporting
4. year: Year of reporting
5. cases: Number of cases for the day
6. deaths: Number of COVID-19 deaths in the day
7. countriesAndTerritories: Name of the country or territory where the case is reported
8. geoId: ISO 3166-1 alpha-2 code of the country or territory
9. countryterritoryCode: ISO 3166-1 alpha-3 code of the country or territory
10. popData2018: Population of the country or territory as at 2018
11. continentExp: Continent of the country or territory

For countryTerritoryCode missing values, 'Anguilla', 'Bonaire, Saint Eustatius and Saba', 'Western_Sahara', 'Cases_on_an_international_conveyance_Japan' and 'Falkland_Islands_(Malvinas)' do not have countryTerritoryCode which can be resolved by assigning their ISO 3166-1 alpha-3 code 'AIA','BES', 'ESH', 'JPG' and "FLK" respectively to them. But for International_conveyance_Japan the alphabeth part of its geoId was used since its ISO 3166-1 alpha-3 code is available.

```python
#Assigned the countryterritoryCode of Bonaire, Saint Eustatius and Saba
bonaire_indices=data.index[data['countriesAndTerritories']=='Bonaire, Saint Eustatius and Saba']
data.loc[bonaire_indices[0]:bonaire_indices[-1],'countryterritoryCode']='BES'
#Assigned the countryterritoryCode of Anguilla
anguilla_indices=data.index[data['countriesAndTerritories']=='Anguilla']
data.loc[anguilla_indices[0]:anguilla_indices[-1],'countryterritoryCode']='AIA'
#Assigned the countryterritoryCode of Falkland_Islands_(Malvinas
falkland_indices=data.index[data['countriesAndTerritories']=='Falkland_Islands_(Malvinas)']
data.loc[falkland_indices[0]:falkland_indices[-1],'countryterritoryCode']='FLK'
#Assigned the countryterritoryCode of Western Sahara
sahara_indices=data.index[data['countriesAndTerritories']=='Western_Sahara']
data.loc[sahara_indices,'countryterritoryCode']='ESH'
#Assigned the countryterritoryCode of Cases_on_an_international_conveyance_Japan
int_japan_indices=data.index[data['countriesAndTerritories']=='Cases_on_an_international_conveyance_Japan']
data.loc[int_japan_indices[0]:int_japan_indices[-1],'countryterritoryCode']='JPG'
```

With respect to popData2018 column the missing values occurred where countries or territories such as Anguilla, Falkland_Islands_(Malvinas), Eriteria, Western_Sahara and Bonaire, Saint Eustatius and Saba were having population. Each country or territory's population was source from https://en.wikipedia.org/wiki/List_of_countries_by_population_(United_Nations) and https://www.internetworldstats.com/list2.htm to assigned to them.

```python
#Assigned the population of Anguilla
anguillapop_indices=data.index[data['countriesAndTerritories']=='Anguilla']
data.loc[anguillapop_indices[0]:anguillapop_indices[-1],'popData2018']=14906
#Assigned the population of Falkland_Islands_(Malvinas)
falklandpop_indices=data.index[data['countriesAndTerritories']=='Falkland_Islands_(Malvinas)']
data.loc[falklandpop_indices[0]:falklandpop_indices[-1],'popData2018']=3234
#Assigned the population of Bonaire, Saint Eustatius and Saba
bonairepop_indices=data.index[data['countriesAndTerritories']=='Bonaire, Saint Eustatius and Saba']
data.loc[bonairepop_indices[0]:bonairepop_indices[-1],'popData2018']=22303
#Assigned the population of Eritrea
eritreapop_indices=data.index[data['countriesAndTerritories']=='Eritrea']
data.loc[eritreapop_indices[0]:eritreapop_indices[-1],'popData2018']=3452786
data.loc[3415:3442,'popData2018']=3452786
#Assigned the countryterritoryCode of Western Sahara
saharapop_indices=data.index[data['countriesAndTerritories']=='Western_Sahara']
data.loc[saharapop_indices,'popData2018']=567402
```

To be able to answer goegraphical questions about COVID-19. There is the need to have columns for geograpical region or area. Classification of countries according to their respective geographical regions is done by using if..elif statements. 

```python
#Initialize lists to classify each affected country to a continental region
'''Extract the countryterritoryCode of each country or territory depending on the continent they belongs'''
africa=data[data['continentExp']== 'Africa'].countryterritoryCode.unique()
asia=data[data['continentExp']== 'Asia'].countryterritoryCode.unique()
americas=data[data['continentExp']== 'America'].countryterritoryCode.unique()
europe=data[data['continentExp']== 'Europe'].countryterritoryCode.unique()
oceania=data[data['continentExp']== 'Oceania'].countryterritoryCode.unique()
```

```python
#Initialize lists to classify each affected country to a geographical region

#AFRICA GEOGRAPHICAL REGIONS
#EASTERN AFRICA
east_africa = ['BDI','COM','DJI','ERI','ETH','KEN','ATF','MDG',
              'MWI','MUS','MYT','MOZ','REU','RWA','SYC','SOM',
              'SSD','TZA','UGA']

#SOUTHERN AFRICA
southern_africa = ['BWA','SWZ','LSO','NAM','ZMB','ZWE','ZAF']

#CENTRAL / MIDDLE AFRICA
central_africa = ['AGO','CMR','CAF','TCD','COD','COG','GNQ','GAB','STP']

#NORTH AFRICA
northern_africa = ['DZA','EGY','LBY','MAR','SDN','TUN','ESH']

#ASIA GOEGRAPHICAL REGIONS
#EASTERN ASIA
east_asia = ['CHN','JPN','MNG','PRK','KOR','HKG','MAC','TWN']

#SOUTHERN ASIA
south_asia = ['AFG','BGD','BTN','MDV','NPL','LKA','PAK','IND','IRN']

#SOUTH-EASTERN ASIA
south_eastern_asia = ['BRN','KHM','IDN','LAO','MYS','MMR','PHL','SGP',
                        'THA','TLS','VNM']

#WESTERN ASIA
western_asia = ['ARM','AZE','ARE','BHR','CYP','KWT','GEO','QAT','PSE','OMN','LBN','ISR','JOR','SYR','YEM',
                        'SAU','IRQ','TUR']


#THE AMERICA GEOGRAPHICAL REGIONS
#NORTH AMERICA
north_america = ['BMU','CAN','GRL','MAF','USA']

#CENTRAL AMERICA
central_america=['BLZ','CRI','SLV','GTM','HND','MEX','NIC','PAN']

#SOUTH AMERICA / LATIN AMERICA
south_america = ['ARG','BOL','BRA','CHL','COL','ECU','FLK','GUF','GUY',
    'PRY','PER','SUR','URY','VEN']

#THE CARRIBEANS
carribeans = ['AIA','ATG','ABW','BHS','BRB','BES','IOT','CYM','CUB','CUW',
    'DMA','DOM','GRD','GLP','HTI','JAM','MTQ','MSR','PRI','BLM',
    'KNA','LCA','VCT','SXM','TTO','TCA','VIR']

#THE EUROPEAN GEOGRAPHICAL REGIONS
#EASTERN EUROPE
eastern_europe = ['BLR','BGR','MDA','ROU','RUS','UKR',
                  'POL','HUN','SVK','CZE']

#WESTERN EUROPE
western_europe = ['AUT','DEU','CHE','FRA','SVN',
                  'NLD','BEL','LUX','MCO','LIE']

#SOUTHERN EUROPE
southern_europe = ['ITA','ESP','GRC','PRT','SRB','HRV','BIH',
                  'ALB','SVN','MKD','MNE','MLT','AND','GIB',
                  'SMR','VAT']
```

```python
#Initialize two list for continental and geographical regions
geo = []
```

```python
#Loop through the column and classify each country to their respective continental and geographical regional zones
for i in data['countryterritoryCode']: 
    if i in africa:
        if i in southern_africa:
            geo.append("Southern Africa")
        elif i in central_africa:
            geo.append("Central Africa")
        elif i in east_africa:
            geo.append("East Africa")
        elif i in northern_africa:
            geo.append("North Africa")
        else:
            geo.append("West Africa")
    elif i in asia:
        if i in east_asia:
            geo.append("Eastern Asia")
        elif i in south_asia:
             geo.append("Southern Asia")
        elif i in south_eastern_asia:
             geo.append("South-eastern Asia")
        elif i in western_asia:
             geo.append("Western Asia")
        else:
            geo.append("Central Asia")
    elif i in oceania:
        geo.append("Oceania")
    elif i in americas:
        if i in north_america:
            geo.append("North America")
        elif i in south_america:
            geo.append("South America")
        elif i in central_america:
            geo.append("Central America")
        else:
            geo.append("Carribeans")
    elif i in europe:
        if i in southern_europe:
            geo.append("Southern Europe")
        elif i in western_europe:
            geo.append("Western Europe")
        elif i in eastern_europe:
            geo.append("Eastern Europe")
        else:
            geo.append("Northern Europe")
    else:
        geo.append("Other")
```

```python
#Create new column for geographical region
data['geographical_region'] = geo
```
